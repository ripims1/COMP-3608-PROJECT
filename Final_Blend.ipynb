```python
import pandas as pd

#1.Load the files
lgbm = pd.read_csv('LGBMsubmission.csv')
nn = pd.read_csv('NNsubmission.csv')
rf = pd.read_csv('RFsubmission.csv')

#2.Convert probabilities to ranks
lgbm_rank = lgbm['Churn'].rank()
nn_rank = nn['Churn'].rank()
rf_rank = rf['Churn'].rank()

#3.Define the weights
w_lgbm = 0.95  #Top performer
w_nn   = 0.04  #Second best
w_rf   = 0.01  #Third best

#4.Calculate the blended rank
blended_rank = (lgbm_rank * w_lgbm) + (nn_rank * w_nn) + (rf_rank * w_rf)

#5.Normalize
final_probs = blended_rank / blended_rank.max()

#6.Create final submission file
final_submission = pd.DataFrame({
    'id': lgbm['id'],
    'Churn': final_probs
})

final_submission.to_csv('final_submission.csv', index=False)

print("Blend complete!")
```
