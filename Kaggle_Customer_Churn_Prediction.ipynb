```python
import pandas as pd

# Importing Training and test datasets
trainData = pd.read_csv('train.csv')
testData = pd.read_csv('test.csv')

# Removing the Id columns for test and training data to avoid being considered during predictions
trainData = trainData.drop('id', axis=1)
testData = testData.drop('id', axis=1)

# Convert Churn data to binary
trainData['Churn'] = trainData['Churn'].map({'Yes': 1, 'No': 0})

# Converting categorical variables into individual dummy variables to be usable with logistic regression
trainData = pd.get_dummies(trainData)
testData = pd.get_dummies(testData)

# Aligning the Training and Test columns to match
trainData, testData = trainData.align(testData, join='left', axis=1, fill_value=0)

# Remove the added Churn column from the Test dataset
testData = testData.drop('Churn', axis=1)

# Finds all boolean columns and stores them as integers 1 and 0 for consistency
booleanColumns = trainData.select_dtypes(include=['bool']).columns
trainData[booleanColumns] = trainData[booleanColumns].astype(int)
testData[booleanColumns] = testData[booleanColumns].astype(int)

# Converting modified datasets to csv and zipping them for easier use
trainData.to_csv('train_processed.csv', index=False)
testData.to_csv('test_processed.csv', index=False)
!zip processed_data.zip train_processed.csv test_processed.csv
```
